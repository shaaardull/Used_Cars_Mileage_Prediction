```python
df["sqrt_of_accln"] = df["acceleration"].apply(np.sqrt)
```

```python
df["disp*horsepower"] = df.apply(lambda x: x["displacement"]*x["horsepower"], axis = 1)
```

# -----------

```python
df = df.drop(["year","origin","acceleration","sqrt_of_accln","disp*horsepower"], axis = 1)
```

```python
df
```

```python
df.corr()
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[["cylinders","displacement","horsepower","weight"]],df["mpg"])
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Let's predict mileage:

```python
def predict_mpg(c,d,h,wt):
    predicted_mpg = reg.predict([[c,d,h,wt]])
    if predicted_mpg >= 0:
        return predicted_mpg
    else: print("Predicted mpg is equal to zero")
```

```python
c = int(input("Enter number of cylinders: "))
d = float(input("Enter displacement in cc: "))
h = float(input("Enter horsepowers produced: "))
wt = float(input("Enter weight of the car: "))
predict_mpg(c,d,h,wt)
```
